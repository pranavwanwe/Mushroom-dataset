### Preprocessing


```python
# lable encoding them using lable encoder 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in df.columns:
    df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

### Splitting data into X and Y

```python
X = df.iloc[:,1:]
Y = df['class']

X.head(5)
```

```python
Y.head()
```

### The data is now encoded 


### Making split for training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
```

```python
X_test.shape
```

```python
X_train.shape
```

### Applying logistic regression


```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression(max_iter=1000)

model_lr.fit(X_train,Y_train)

Y_pred = model_lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print("confusion matrix", confusion_matrix(Y_test,Y_pred))
print("classification report:", classification_report(Y_test,Y_pred))
print("accuracy score:", accuracy_score(Y_test,Y_pred)*100)
```

### With logistic regression we got 95.32% accuracy, Let's now try SVC

```python
from sklearn.svm import SVC

model_svc = SVC()

model_svc.fit(X_train,Y_train)

Y_pred = model_svc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print("confusion matrix", confusion_matrix(Y_test,Y_pred))
print("classification report:", classification_report(Y_test,Y_pred))
print("accuracy score:", accuracy_score(Y_test,Y_pred)*100)
```

### With SVC we have got 99.13% of accuracy, now let's try KNN 

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier()

model_knn.fit(X_train,Y_train)

Y_pred = model_knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print("confusion matrix", confusion_matrix(Y_test,Y_pred))
print("classification report:", classification_report(Y_test,Y_pred))
print("accuracy score:", accuracy_score(Y_test,Y_pred)*100)
```

### With KNN we have got some increase in accuracy that is 99.81%, Let's try decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier()

model_dt.fit(X_train,Y_train)

Y_pred = model_dt.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print("confusion matrix", confusion_matrix(Y_test,Y_pred))
print("classification report:", classification_report(Y_test,Y_pred))
print("accuracy score:", accuracy_score(Y_test,Y_pred)*100)
```

### With Decision tree classifier we have got 100% accuracy
